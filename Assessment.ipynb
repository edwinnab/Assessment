```python
#merge the three dataset into one final dataset
first_Two = pd.merge(boxPlacement, wetHarvest, on="form.case.@case_id")
final_dataset = pd.merge(first_Two, dryHarvest, on="form.case.@case_id")
final_dataset.info()
#the final dataset has 86columns and 52rows
```

```python
#display the first five rows of final dataset
final_dataset.head()
```

```python
#clean the dataset
#find the columns with empty rows
#the phonenumber2 column has 52 filled in records
#it shows that most farmers do not have alternative phone numbers
null_columns = final_dataset.columns[final_dataset.isnull().any()]
final_dataset[null_columns].isnull().sum()
```

```python
#drop the column as it is incomplete
final_dataset=final_dataset.dropna(axis='columns')
final_dataset.info()
```

```python
#check the box measurements
final_dataset["Total_Measures"] = final_dataset["Length of first box- enter -99 if you could not place the box"] * final_dataset["width of the first box, enter -99 if you could not place the box"]
final_dataset.head()
#total_measures colum indicates the normal measurements of boxes 
```

```python
#we well show the rows with crazy measurements
final_dataset[final_dataset.Total_Measures > 40]
```

```python
#we well show the rows with crazy measurements
final_dataset[final_dataset.Total_Measures < 40]
```

```python
final_dataset["Total_Dry"] = final_dataset["Dry weight of Soybeans from the First bag"] + final_dataset["Dry weight of Soybeans from the Second bag"]
final_dataset.head()
```

```python
#add a new column that calculate the total wet weights
final_dataset["Total_wet"] = final_dataset['weight of soybeans in first box Remember, if it was planted and FAILED you need to enter 0 for the weight. If the farmer harvested a crop without you, indicate -99 for that crop which means "harvested without me".'] + final_dataset['weight of Soybeans in the second box Remember, if it was planted and FAILED you need to enter 0 for the weight. If the farmer harvested a crop without you, indicate -99 for that crop which means "harvested without me".']
final_dataset
```

```python
##compares the dry harvests with the wet harvests 
final_dataset["compared"] =final_dataset["Total_Dry"] > final_dataset["Total_wet"]
final_dataset
```

```python
##adds the distance between the crops before finding averages
final_dataset["Distance"]=final_dataset['Go to the FIRST row in the FIRST box and take the measurement between the FIRST living plant and the SECOND living plant. Record your answer here in centimeters'] + final_dataset['Go to the FIRST row in the SECOND box and take the measurement between the SECOND living plant and the THIRD living plant. Record your answer here in centimeters']
final_dataset
```

```python
#calculate the average row spacing in metres
final_dataset["average_distance"]=(final_dataset["Distance"]/2)/100
final_dataset
```

```python
#compute average yields per mt/ha
final_dataset["average_yields(mt/ha)"] = (final_dataset["Total_Dry"]*10000)/(2*40*final_dataset["average_distance"])
final_dataset
```

```python
#creating a spatial distribution
import matplotlib.pyplot as plt
```

```python
#create a spatial distribution to see the areas with high dry harvests
plt.figure(figsize = (10,8))
plt.scatter(final_dataset.Longitude_bp, final_dataset.Latitude_bp, c=final_dataset.Total_Dry, cmap = "hot", s=1)
plt.colorbar().set_label("Total_Dry_Weights depending_on_Area", fontsize=14)
plt.xlabel("Longitude", fontsize=14)
plt.ylabel("Latitude", fontsize=14)
plt.title("Spatial Distribution of Harvests", fontsize=17)
plt.show()
```
